## 3. Preparación de los datos.

```python
# Los nombres de las columnas empiezan con mayúsculas, pon todos los nombres de las columnas en minúsculas.

# Creo el diccionario con los nombres sustitutos

diccionario_nombres = {}

for col in columnas:
    diccionario_nombres[col]= col.lower()

df.rename(columns= diccionario_nombres, inplace=True)

df.head(2)
```

 o Haciendo el análisis exploratorio te deberías haber dado cuenta de que algunas de las 
variables (danceability, energy, key, loudness, speechiness, acousticness, 
instrumentalness) no son del tipo que deberían. Esto es debido a que los decimales 
están establecidos como comas y no con puntos. Crea una función que nos permita 
cambiar esas comas por puntos para que los datos tengan el tipo correcto.
 ￿ Después de haber hecho los cambios, chequea los tipos de datos. ¿Son ya del 
tipo correcto? En caso de que no, crea otra función o modifica la anterior para 
que sean de tipo float.

```python
cols_a_float = ['danceability', 'energy', 'key', 'loudness', 'speechiness', 'acousticness', 
'instrumentalness']

# Creo una función para castear 

def cadena_a_float (cadena):
    try:
        resultado = float(cadena.replace(",", "."))
    except:
        resultado = np.nan  # Si encuentra un nulo, devuelve un nan.

    return resultado

# Aplico la función a todas las columnas que son object y deberían ser float

for col in cols_a_float:
    df[col] = df[col].apply(cadena_a_float)

# Compruebo los resultados

df['danceability'].dtype
```

 o Algunos de los artistas tienen símbolos raros, en concreto el símbolo $. Crea una función 
que los reemplaze pos "s".

```python
display(df['artist'].unique())
```

```python
def dolar_a_s (cadena):
    try:
        return cadena.replace('$', 's')
    except:
        return np.nan
    
for artist in df['artist']:
    df['artist'] = df['artist'].apply(dolar_a_s)


# El nombre de los artistas con $ no aparece solo en la columna de los artistas, sino también en album, pista, título, channel y description.
# Aplicamos nuestra función a esas columnas

columnas_con_nombre_de_artista = ['album', 'title', 'track', 'channel', 'description']

for col in columnas_con_nombre_de_artista:
    df[col] = df[col].apply(dolar_a_s)
```

```python
# Filtramos por condición

condicion_album = df['album_type'] == 'album'
condicion_single = df['album_type'] == 'single'
condicion_compilation = df['album_type'] == 'compilation'

df_album = df[condicion_album]
df_single = df[condicion_single]
df_compilation = df[condicion_compilation]

print(f"El DF de albumes tiene {df_album.shape[0]} canciones")
print(f"El DF de singles tiene {df_single.shape[0]} canciones")
print(f"El DF de recopilatorios tiene {df_compilation.shape[0]} canciones")
```

 o Usando el dataframe de los albumes que hemos creado en el ejercicio anterior. ¿Cuál es 
la media y la desviación estándar de "danceability", "acousticness" y "speechiness" de 
cada artista?. Muesrtra los resultados en tres dataframes diferentes, uno para cada 
métrica ("danceability", "acousticness" y "speechiness"). Ordena los resultados de mayor 
a menor en base a la media.

```python

# Agrupamos por artista y sacamos sus métricas

df_danceability = df_album.groupby("artist")['danceability'].agg(['mean', 'std']).reset_index().sort_values(by='mean', ascending=False)
df_acousticness = df_album.groupby("artist")['acousticness'].agg(['mean', 'std']).reset_index().sort_values(by='mean', ascending=False)
df_speechiness = df_album.groupby("artist")['speechiness'].agg(['mean', 'std']).reset_index().sort_values(by='mean', ascending=False)

display(df_danceability,df_acousticness, df_speechiness) 
```

 o Se quiere estudiar si existe una relación entre la cantidad de visitas que reciben los 
videos de música de un artista en una plataforma de streaming y la cantidad de likes que 
estos videos obtienen. Por lo tanto, queremos contestar a la siguiente pregunta: ¿Son 
aquellos artistas que tienen más visitas los que más likes tienen? Para solucionar este 
ejercicio deberéis:

 ￿ Agrupa por artista y calcular la media de visitas y de likes.
 ￿ Quedate con los 10 artistas que más visitas han tenido.
 ￿ Quedate con los 10 artistas que más likes han tenido.

```python
# Agrupamos por artista y sacamos los promedios de visitas y likes.

df_relacion_visitas_likes_orden_v = df.groupby('artist')[['views','likes']].mean().reset_index().sort_values('views', ascending=False) # Ordenamos por views 
df_relacion_visitas_likes_orden_l = df.groupby('artist')[['views','likes']].mean().reset_index().sort_values('views', ascending=False) # Ordenamos por likes 

# Nos quedamos con los 10 primeros de cada tabla ordenada

diez_artistas_con_mas_views = df_relacion_visitas_likes_orden_v['artist'].head(10).reset_index()
diez_artistas_con_mas_likes = df_relacion_visitas_likes_orden_l['artist'].head(10).reset_index()

display(diez_artistas_con_mas_views, diez_artistas_con_mas_likes)

# Comprobamos que son los mismos 10. Deducimos hay una relación directa: los que más visitas tienen son los que más likes tienen.
```
